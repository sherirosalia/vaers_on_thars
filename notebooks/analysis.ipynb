```python
data=az_complete_left_join
data['SEX_FLOAT']=data['SEX'].apply(lambda gender: 0 if gender=='F' else 1)
# .map(lambda name: name.lower().replace("'",""))
```

```python
data['VAX_SITE_int']=data['VAX_SITE'].astype('category').cat.codes
```

```python
# columns = ['SYMPTOM2','SYMPTOMVERSION2','SYMPTOM3',	'SYMPTOMVERSION3','SYMPTOM4','SYMPTOMVERSION4','SYMPTOM5','SYMPTOMVERSION5']
```

```python
# one hot encoding have a column for each symptom 
from sklearn.preprocessing import OneHotEncoder
```

```python
symptom1 = pd.get_dummies(data[['SYMPTOM1']],prefix=['sym'], columns = ['SYMPTOM1'], drop_first=True)
symptom2 = pd.get_dummies(data[['SYMPTOM2']],prefix=['sym'], columns = ['SYMPTOM2'], drop_first=True)
symptom3 = pd.get_dummies(data[['SYMPTOM3']],prefix=['sym'], columns = ['SYMPTOM3'], drop_first=True)
symptom4 = pd.get_dummies(data[['SYMPTOM4']],prefix=['sym'], columns = ['SYMPTOM4'], drop_first=True)
```

```python
# symptom1 + symptom2 + symptom3 + symptom4
symp=symptom1.add(symptom2, fill_value=0).add(symptom3, fill_value=0).add(symptom4, fill_value=0)
symp
```

```python
columns=['AGE_YRS','SEX_FLOAT','VAX_SITE_int']
symp[['AGE_YRS','SEX_FLOAT','VAX_SITE_int']]=data[['AGE_YRS','SEX_FLOAT','VAX_SITE_int']]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
X = symp.fillna(-1)
y = data['VAX_MANU']
# X is a dataframe and y a series - correct way to run
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=42)
```

```python
clf = RandomForestClassifier(max_depth=2, random_state=0)
clf.fit(X_train, y_train)
```

```python
clf.predict(X_test)
```

```python
clf.score(X_train, y_train)
```
